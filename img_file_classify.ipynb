```python
hasGo = True
noGo = True
for f in list:
    oldPath = path+"img_align_celeba/"+f
    
    noArray = noLine.split()
    if(len(noArray)<1):
        noGo = False
    hasArray = hasLine.split()
    if(len(hasArray)<1):
        hasGo = False
        
    if (noGo and (f==noArray[0])):
        newPath = path+"noGlass/"
        shutil.copy(oldPath,newPath)
        noLine = nof.readline()
        
    if(hasGo and (f==hasArray[0])):
        newPath = path+"hasGlass/"
        shutil.copy(oldPath,newPath)
        hasLine = hasf.readline()
```
